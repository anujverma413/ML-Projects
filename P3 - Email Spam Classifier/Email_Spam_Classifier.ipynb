```python
df1['spam']=df1['Category'].apply(lambda x: 1 if x=='spam' else 0)
df1.head()
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
len(df1)
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
def edit_message(message):
    edited_message = re.sub('[^a-zA-Z]',' ',message)
    edited_message = edited_message.lower()
    edited_message = edited_message.split()
    edited_message = [ps.stem(word) for word in edited_message if not word in stopwords.words('english')]
    edited_message = ' '.join(edited_message)
    return edited_message
for i in range(0, len(df1)):
    edited_message = edit_message(df1['Message'][i])
    corpus.append(edited_message)
```

```python
corpus
```

```python
words = []
for message in corpus:
    words = words + message.split()
words
```

```python
len(words)
```

```python
Bag_of_words = set(words)
len(Bag_of_words)
```

```python
# len(dictionary)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
CV = CountVectorizer(vocabulary = Bag_of_words)
X = CV.fit_transform(corpus).toarray()
```

```python
Y = df1.spam
```

```python
X
```

```python
X.shape, Y.shape 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 10)
```

```python
X.shape, X_train.shape, X_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train,Y_train)
```

```python
X_train_score = model.score(X_train, Y_train)
X_train_score
```

```python
X_test_score = model.score(X_test, Y_test)        
X_test_score                                      
```

```python
from sklearn.metrics import confusion_matrix
Y_pred = model.predict(X_test)
conf_matrix = confusion_matrix(Y_test, Y_pred)
conf_matrix    
# (1,1)-True Negative, (1,2)-False Positive, (2,1)-False Negative, (2,2)-True Positive
```

```python
# Using Pipeline Methos
```

```python
X = df1.Message
Y = df1.spam
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 10)
```

```python
from sklearn.pipeline import Pipeline
classifier = Pipeline([
    ('vectorizer',  CountVectorizer(vocabulary = Bag_of_words)),
    ('nb', MultinomialNB())
])
```

```python
classifier.fit(X_train, Y_train)
```

```python
classifier.score(X_train, Y_train)
```

```python
classifier.score(X_test, Y_test)
```

```python
emails = ['This Diwali, a huge sale is out. 50% discount on each item. Place your order before the stock is empty.',
           'I saw your profile on Linkedin. You seem to be fit for the job. Would you like to join from next Monday? ']
classifier.predict(emails)
```
